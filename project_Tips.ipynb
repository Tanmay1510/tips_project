```python
#we have imported libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# we will load dataset
tips = sns.load_dataset('tips')
print("Dataset loaded")
```

```python
#added new column tip percentage
tips['tip_percent']=(tips['tip']/tips['total_bill'])*100
```

```python
#initialize weekdays and wekends
tips['day_type.type'] = np.where(tips['day'].isin(['Sat', 'Sun']), 'Weekend', 'Weekday')
```

```python
#compared
weekend = tips[tips['day_type.type']=='Weekend']
weekday = tips[tips['day_type.type']== 'Weekday']
```

```python
#avg value
print("Avg bill=")
print("Weekend:", weekend['total_bill'].mean())
print("Weekday:", weekday['total_bill'].mean())
```

```python
print("Avg tip=")
print("Weekend:", weekend['tip'].mean())
print("Weekday:", weekday['tip'].mean())
```

```python
print("avg %=")
print("Weekend:", weekend['tip_percent'].mean())
print("Weekday:", weekday['tip_percent'].mean())
```

```python
# comparision of total bill
plt.figure(figsize=(8,5))
sns.boxplot(x='day_type.type', y='total_bill', data=tips)
plt.title('Total Bill: Weekend vs Weekday')
plt.show()
```

```python
#comparision of tip
plt.figure(figsize=(8,5))
sns.boxplot(x='day_type.type', y='tip', data=tips)
plt.title('Tip Amount: Weekend vs Weekday')
plt.show()
```

```python
#tip %
plt.figure(figsize=(8,5))
sns.boxplot(x='day_type.type', y='tip_percent', data=tips)
plt.title('Tip Percentage: Weekend vs Weekday')
plt.show()
```

```python
#total bill distribution of day by day
plt.figure(figsize=(8,5))
sns.barplot(x='day', y='total_bill', data=tips, estimator=np.mean)
plt.title('Average Total Bill by Day')
plt.show()
```

```python
#summary
print("Final Summary:")
if weekend['tip_percent'].mean() > weekday['tip_percent'].mean():
    print("People tip more (in %) on weekends!")
else:
    print("People tip more in percentage on weekdays")

if weekend['total_bill'].mean() > weekday['total_bill'].mean():
    print("People spend more on weekends")
else:
    print("People spend more on weekdays")
```
